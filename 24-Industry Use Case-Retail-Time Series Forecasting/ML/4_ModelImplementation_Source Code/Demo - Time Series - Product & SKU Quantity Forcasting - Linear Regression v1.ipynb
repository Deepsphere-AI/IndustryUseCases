# Product Level Forecasting

```python
#********************************************************************************

## Step 1: INI File Configuration

#********************************************************************************

import configparser
import os
vAR_Config = configparser.ConfigParser(allow_no_value=True)

vAR_INI_FILE_PATH = os.getenv('TIME_SERIES_INI_ML')
#print(vAR_INI_FILE_PATH)

vAR_Config.read(vAR_INI_FILE_PATH)

vAR_Data = vAR_Config.sections()

vAR_Config.sections()

vAR_Training_Data = vAR_Config['FILE PATH']['TRAINING_DATA']
#print(vAR_Training_Data)

vAR_Training_Data_Excel_Worsheet = vAR_Config['FILE PATH']['TRAINING_DATA_EXCEL_WORKSHEET']
#print(vAR_Training_Data_Excel_Worsheet)

vAR_Test_Data = vAR_Config['FILE PATH']['TEST_DATA']
#print(vAR_Test_Data)

vAR_Test_Data_Excel_Worsheet = vAR_Config['FILE PATH']['TEST_DATA_EXCEL_WORKSHEET']
#print(vAR_Test_Data_Excel_Worsheet)

vAR_FORECAST_PRODUCT_QUANTITY = vAR_Config['FILE PATH']['FORECAST_PRODUCT_QUANTITY']
#print(vAR_FORECAST_PRODUCT_QUANTITY)

vAR_FORECAST_SKU_QUANTITY = vAR_Config['FILE PATH']['FORECAST_SKU_QUANTITY']
#print(vAR_FORECAST_SKU_QUANTITY)

vAR_FORECAST_PRODUCT_QUANTITY_ALL = vAR_Config['FILE PATH']['FORECAST_PRODUCT_QUANTITY_ALL']
#print(vAR_FORECAST_PRODUCT_QUANTITY_ALL)

vAR_FORECAST_SKU_QUANTITY_ALL = vAR_Config['FILE PATH']['FORECAST_SKU_QUANTITY_ALL']
#print(vAR_FORECAST_SKU_QUANTITY_ALL)

vAR_FORECAST_PRODUCT_SKU_QUANTITY_ALL = vAR_Config['FILE PATH']['FORECAST_PRODUCT_SKU_QUANTITY_ALL']
#print(vAR_FORECAST_PRODUCT_SKU_QUANTITY_ALL)

vAR_6_MONTHS_ROLLING_FORECAST = vAR_Config['FILE PATH']['6_MONTHS_ROLLING_FORECAST']
#print(vAR_6_MONTHS_ROLLING_FORECAST)

vAR_SALES_REVENUE = vAR_Config['FILE PATH']['SALES_REVENUE']
#print(vAR_SALES_REVENUE)

vAR_FORECAST_COLUMN = vAR_Config['OUTPUT']['FORECAST_COLUMN']
#print(vAR_FORECAST_COLUMN)

vAR_FORECAST_COLUMN1 = vAR_Config['OUTPUT']['FORECAST_COLUMN1']
#print(vAR_FORECAST_COLUMN1)


#********************************************************************************
    
## Step 2: Importing the Required Libraries

#********************************************************************************

import pandas as vAR_pd

import numpy as vAR_np

from sklearn.linear_model import LinearRegression

from sklearn.preprocessing import LabelEncoder


#********************************************************************************
    
## Step 3: Importing the Training Data

#********************************************************************************

vAR_INPUT_DATA = vAR_pd.read_excel(vAR_Training_Data,sheet_name=vAR_Training_Data_Excel_Worsheet)

vAR_INPUT_DATA = vAR_pd.DataFrame(vAR_INPUT_DATA)

vAR_INPUT_DATA.head()

vAR_TRAINING_DATA = vAR_INPUT_DATA

vAR_TRAINING_DATA.head()

vAR_le = LabelEncoder()


vAR_Region_Conversion = vAR_le.fit_transform(vAR_TRAINING_DATA.iloc[:,0])

vAR_Region_Conversion_df = vAR_pd.DataFrame(vAR_Region_Conversion,columns={'Region_Converted'})


vAR_County_Conversion = vAR_le.fit_transform(vAR_TRAINING_DATA.iloc[:,1])

vAR_County_Conversion_df = vAR_pd.DataFrame(vAR_County_Conversion,columns={'County_Converted'})


vAR_State_Conversion = vAR_le.fit_transform(vAR_TRAINING_DATA.iloc[:,2])

vAR_State_Conversion_df = vAR_pd.DataFrame(vAR_State_Conversion,columns={'State_Converted'})


vAR_City_Conversion = vAR_le.fit_transform(vAR_TRAINING_DATA.iloc[:,3])

vAR_City_Conversion_df = vAR_pd.DataFrame(vAR_City_Conversion,columns={'City_Converted'})


vAR_Customer_Conversion = vAR_le.fit_transform(vAR_TRAINING_DATA.iloc[:,4])

vAR_Customer_Conversion_df = vAR_pd.DataFrame(vAR_Customer_Conversion,columns={'Customer_Converted'})


vAR_Product_Family_Conversion = vAR_le.fit_transform(vAR_TRAINING_DATA.iloc[:,5])

vAR_Product_Family_Conversion_df = vAR_pd.DataFrame(vAR_Product_Family_Conversion,columns={'Product_family_Converted'})


vAR_Product_Group_Conversion = vAR_le.fit_transform(vAR_TRAINING_DATA.iloc[:,6])

vAR_Product_Group_Conversion_df = vAR_pd.DataFrame(vAR_Product_Group_Conversion,columns={'Product Group_Converted'})


vAR_Product_Conversion = vAR_le.fit_transform(vAR_TRAINING_DATA.iloc[:,7])

vAR_Product_Conversion_df = vAR_pd.DataFrame(vAR_Product_Conversion,columns={'Product_Converted'})


vAR_SKU_Conversion = vAR_le.fit_transform(vAR_TRAINING_DATA.iloc[:,8])

vAR_SKU_Conversion_df = vAR_pd.DataFrame(vAR_SKU_Conversion,columns={'SKU_Converted'})


# Attached the Converted Numerical Data to the main dataframe

vAR_df1 = vAR_pd.DataFrame(columns=['Region','County','State','City','Customer','Product Family','Product Group','Product','SKU'])

vAR_df1['Region'] = vAR_Region_Conversion

vAR_df1['County'] = vAR_County_Conversion

vAR_df1['State'] = vAR_State_Conversion

vAR_df1['City'] = vAR_City_Conversion

vAR_df1['Customer'] = vAR_Customer_Conversion

vAR_df1['Product Family'] = vAR_Product_Family_Conversion

vAR_df1['Product Group'] = vAR_Product_Group_Conversion

vAR_df1['Product'] = vAR_Product_Conversion

vAR_df1['SKU'] = vAR_SKU_Conversion

vAR_df1['Year'] = vAR_TRAINING_DATA['Year']

vAR_df1['Month'] = vAR_TRAINING_DATA['Month']

#vAR_df1['Time'] = vAR_TRAINING_DATA['Time']

vAR_TRAINING_DATA_PRODUCT_QUANTITY_FEATURES = vAR_df1[['Region', 'County', 'State', 'City', 'Customer', 'Product Family','Product Group', 'Product', 'SKU', 'Year', 'Month']]

vAR_TRAINING_DATA_PRODUCT_QUANTITY_FEATURES

vAR_TRAINING_DATA_PRODUCT_QUANTITY_LABEL = vAR_TRAINING_DATA['Quantity_Product']


#********************************************************************************
    
## Step 4: Importing the Linear Regression Model

#********************************************************************************

vAR_MODEL = LinearRegression()


#********************************************************************************
    
## Step 5: Training the Model

#********************************************************************************

vAR_MODEL_TRAINING_PRODUCT_QUANTITY = vAR_MODEL.fit(vAR_TRAINING_DATA_PRODUCT_QUANTITY_FEATURES,vAR_TRAINING_DATA_PRODUCT_QUANTITY_LABEL)


#********************************************************************************
    
## Step 6: Import the Test Data

#********************************************************************************

vAR_TEST_DATA = vAR_pd.read_excel(vAR_Test_Data,sheet_name=vAR_Test_Data_Excel_Worsheet)

vAR_TEST_DATA = vAR_pd.DataFrame(vAR_TEST_DATA)

vAR_TEST_DATA.head()


vAR_Region_Conversion = vAR_le.fit_transform(vAR_TEST_DATA.iloc[:,0])

vAR_County_Conversion = vAR_le.fit_transform(vAR_TEST_DATA.iloc[:,1])

vAR_State_Conversion = vAR_le.fit_transform(vAR_TEST_DATA.iloc[:,2])

vAR_City_Conversion = vAR_le.fit_transform(vAR_TEST_DATA.iloc[:,3])

vAR_Customer_Conversion = vAR_le.fit_transform(vAR_TEST_DATA.iloc[:,4])

vAR_Product_Family_Conversion = vAR_le.fit_transform(vAR_TEST_DATA.iloc[:,5])

vAR_Product_Group_Conversion = vAR_le.fit_transform(vAR_TEST_DATA.iloc[:,6])

vAR_Product_Conversion = vAR_le.fit_transform(vAR_TEST_DATA.iloc[:,7])

vAR_SKU_Conversion = vAR_le.fit_transform(vAR_TEST_DATA.iloc[:,8])


# Attached the Converted Numerical Data to the main dataframe

vAR_df2 = vAR_pd.DataFrame(columns=['Region','County','State','City','Customer','Product Family','Product Group','Product','SKU'])

vAR_df2['Region'] = vAR_Region_Conversion

vAR_df2['County'] = vAR_County_Conversion

vAR_df2['State'] = vAR_State_Conversion

vAR_df2['City'] = vAR_City_Conversion

vAR_df2['Customer'] = vAR_Customer_Conversion

vAR_df2['Product Family'] = vAR_Product_Family_Conversion

vAR_df2['Product Group'] = vAR_Product_Group_Conversion

vAR_df2['Product'] = vAR_Product_Conversion

vAR_df2['SKU'] = vAR_SKU_Conversion

vAR_df2['Year'] = vAR_TRAINING_DATA['Year']

vAR_df2['Month'] = vAR_TRAINING_DATA['Month']

vAR_df2.head()

vAR_TEST_DATA_PRODUCT_QUANTITY_FEATURES = vAR_df2[['Region', 'County', 'State', 'City', 'Customer', 'Product Family','Product Group', 'Product', 'SKU', 'Year', 'Month']]


#********************************************************************************
    
## Step 7: Run Forecasting

#********************************************************************************

vAR_MODEL_TEST_PRODUCT_QUANTITY = vAR_MODEL.predict(vAR_TEST_DATA_PRODUCT_QUANTITY_FEATURES)

vAR_MODEL_TEST_PRODUCT_QUANTITY = vAR_pd.DataFrame(vAR_MODEL_TEST_PRODUCT_QUANTITY,columns=['Forecasted Product Quantity']).astype(int)

vAR_TEST_DATA['Forecasted Product Quantity'] = vAR_MODEL_TEST_PRODUCT_QUANTITY

vAR_MODEL_OUTCOME_FORECASTED_QUANTITY_PRODUCT = vAR_TEST_DATA


#********************************************************************************
    
## Step 8: Write the Model Outcome to a file

#********************************************************************************


vAR_TEST_DATA = vAR_TEST_DATA.to_excel(vAR_FORECAST_PRODUCT_QUANTITY_ALL,index=False) 

vAR_TEST_DATA = vAR_pd.read_excel(vAR_FORECAST_PRODUCT_QUANTITY_ALL)

vAR_TEST_DATA.head()

```

# 6 Months Rolling Forecast

```python
vAR_FORECAST_PRODUCT_SKU_QUANTITY_ALL = vAR_pd.read_excel(vAR_FORECAST_PRODUCT_SKU_QUANTITY_ALL)
                                     
vAR_ROLLING_FORECAST = vAR_FORECAST_PRODUCT_SKU_QUANTITY_ALL[vAR_FORECAST_PRODUCT_SKU_QUANTITY_ALL['Year']==2020]

vAR_ROLLING_FORECAST_JUN = vAR_ROLLING_FORECAST[vAR_FORECAST_PRODUCT_SKU_QUANTITY_ALL['Month']=='Jun']

vAR_ROLLING_FORECAST_JUN.reset_index()

vAR_ROLLING_FORECAST_JUL = vAR_ROLLING_FORECAST[vAR_FORECAST_PRODUCT_SKU_QUANTITY_ALL['Month']=='Jul']

vAR_ROLLING_FORECAST_JUL.reset_index()

vAR_ROLLING_FORECAST_AUG = vAR_ROLLING_FORECAST[vAR_FORECAST_PRODUCT_SKU_QUANTITY_ALL['Month']=='Aug']

vAR_ROLLING_FORECAST_SEP = vAR_ROLLING_FORECAST[vAR_FORECAST_PRODUCT_SKU_QUANTITY_ALL['Month']=='Sep']

vAR_ROLLING_FORECAST_OCT = vAR_ROLLING_FORECAST[vAR_FORECAST_PRODUCT_SKU_QUANTITY_ALL['Month']=='Oct']

vAR_ROLLING_FORECAST_NOV = vAR_ROLLING_FORECAST[vAR_FORECAST_PRODUCT_SKU_QUANTITY_ALL['Month']=='Nov']

vAR_ROLLING_FORECAST_FINAL = vAR_pd.concat([vAR_ROLLING_FORECAST_JUN,vAR_ROLLING_FORECAST_JUL,vAR_ROLLING_FORECAST_AUG,vAR_ROLLING_FORECAST_SEP,vAR_ROLLING_FORECAST_OCT,vAR_ROLLING_FORECAST_NOV])

vAR_ROLLING_FORECAST_FINAL

vAR_ROLLING_FORECAST = vAR_ROLLING_FORECAST_FINAL.to_excel(vAR_6_MONTHS_ROLLING_FORECAST,index=False)

vAR_ROLLING_FORECAST = vAR_pd.read_excel(vAR_6_MONTHS_ROLLING_FORECAST)

vAR_ROLLING_FORECAST.head()
```
