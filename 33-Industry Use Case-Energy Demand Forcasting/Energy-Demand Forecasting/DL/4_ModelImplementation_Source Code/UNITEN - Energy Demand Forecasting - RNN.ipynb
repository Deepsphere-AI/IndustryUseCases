# ENERGY DEMAND FORECASTING USING RNN

```python
#********************************************************************************
    
## Step 1: Importing the Required Libraries

#********************************************************************************

import os

import yaml

vAR_yaml_file_path = 'C:/AI/Energy Demand Forecasting/DL/YAML/'

for vAR_path in os.listdir(vAR_yaml_file_path):
   
    vAR_final_path = os.path.join(vAR_yaml_file_path,'ENERGY_DEMAND_FORECASTING.yaml')
   
    with open (vAR_final_path) as vAR_file:
      
      config = yaml.safe_load(vAR_file)

## Reading the Training Data Path

vAR_Training_Data_Path = config.get("TRAINING_DATA_PATH")
   
#print(vAR_Training_Data_Path)

vAR_Test_Data_Path = config.get("TEST_DATA_PATH")

#print(vAR_Test_Data_Path)

vAR_Model_Outcome = config.get("MODEL_OUTCOME")

#print(vAR_Model_Outcome)

vAR_Model_Outcome1 = config.get("MODEL_OUTCOME_ALL_DIMENSIONS")

#print(vAR_Model_Outcome1)

vAR_Model_Outcome_6Months_Rolling_Forecast = config.get("MODEL_OUTCOME_6MONTHS_ROLLLING_FORECAST")

#print(vAR_Model_Outcome_6Months_Rolling_Forecast)


#********************************************************************************
    
## Step 2: Importing the Required Libraries

#********************************************************************************

import pandas as vAR_pd

import numpy as vAR_np

from sklearn.preprocessing import MinMaxScaler

from sklearn.preprocessing import LabelEncoder

from sklearn.preprocessing import MinMaxScaler

from keras.models import Sequential

from keras.layers import Dense

from keras.layers import LSTM

from keras.layers import Dropout


#********************************************************************************
    
## Step 3: Importing the Training Data

#********************************************************************************

vAR_INPUT_DATA = vAR_pd.read_excel(vAR_Training_Data_Path)

vAR_INPUT_DATA = vAR_pd.DataFrame(vAR_INPUT_DATA)

vAR_INPUT_DATA.head()

vAR_TRAINING_DATA = vAR_INPUT_DATA

#vAR_TRAINING_DATA.head()

vAR_le = LabelEncoder()


#vAR_Region_Conversion = vAR_le.fit_transform(vAR_TRAINING_DATA.iloc[:,0])

vAR_Region_Conversion_df = vAR_pd.DataFrame(vAR_Region_Conversion,columns={'Region_Converted'})


#vAR_County_Conversion = vAR_le.fit_transform(vAR_TRAINING_DATA.iloc[:,1])

vAR_County_Conversion_df = vAR_pd.DataFrame(vAR_County_Conversion,columns={'County_Converted'})


#vAR_State_Conversion = vAR_le.fit_transform(vAR_TRAINING_DATA.iloc[:,2])

#vAR_State_Conversion_df = vAR_pd.DataFrame(vAR_State_Conversion,columns={'State_Converted'})

#vAR_State_Conversion_df


#vAR_City_Conversion = vAR_le.fit_transform(vAR_TRAINING_DATA.iloc[:,3])

#vAR_City_Conversion_df = vAR_pd.DataFrame(vAR_City_Conversion,columns={'City_Converted'})

#vAR_City_Conversion_df


# Attached the Converted Numerical Data to the main dataframe

vAR_df1 = vAR_pd.DataFrame(columns=['Region','County','State','City', 'Year','Month','Energy Consumption in MW'])

vAR_df1['Region'] = vAR_Region_Conversion

vAR_df1['County'] = vAR_County_Conversion

vAR_df1['State'] = vAR_State_Conversion

vAR_df1['City'] = vAR_City_Conversion

vAR_df1['Year'] = vAR_TRAINING_DATA['Year']

vAR_df1['Month'] = vAR_TRAINING_DATA['Month']

vAR_df1['Energy Consumption in MW'] = vAR_TRAINING_DATA['Energy Consumption in MW']

vAR_TRAINING_DATA_PRODUCT_QUANTITY_FEATURES = vAR_df1[['Region', 'County', 'State', 'City', 'Year','Month','Energy Consumption in MW']]

vAR_TRAINING_DATA_PRODUCT_QUANTITY_LABEL = vAR_TRAINING_DATA.iloc[:,7].values

vAR_TRAINING_DATA_PRODUCT_QUANTITY_LABEL

#vAR_TRAINING_DATA_PRODUCT_QUANTITY_FEATURES = vAR_TRAINING_DATA_PRODUCT_QUANTITY_FEATURES.astype('float32')

vAR_TRAINING_DATA_PRODUCT_QUANTITY_LABEL = vAR_TRAINING_DATA_PRODUCT_QUANTITY_LABEL.reshape(-1,1)

vAR_scaler = MinMaxScaler(feature_range=(0, 1))

vAR_TRAINING_DATA_PRODUCT_QUANTITY_SCALED_LABEL = vAR_scaler.fit_transform(vAR_TRAINING_DATA_PRODUCT_QUANTITY_LABEL)

vAR_TRAINING_DATA_PRODUCT_QUANTITY_LABEL = vAR_TRAINING_DATA['Energy Consumption in MW']

vAR_TRAINING_DATA_PRODUCT_QUANTITY_FEATURES_train = []

vAR_TRAINING_DATA_PRODUCT_QUANTITY_LABEL_train = []

vAR_X_train = []

vAR_y_train = []

for i in range(60, 6480):
    
    vAR_X_train.append(vAR_TRAINING_DATA_PRODUCT_QUANTITY_SCALED_LABEL[i-60:i, 0])
    
    vAR_y_train.append(vAR_TRAINING_DATA_PRODUCT_QUANTITY_SCALED_LABEL[i, 0])

vAR_X_train, vAR_y_train = vAR_np.array(vAR_X_train), vAR_np.array(vAR_y_train)

vAR_X_train = vAR_np.reshape(vAR_X_train, (vAR_X_train.shape[0], vAR_X_train.shape[1], 1))

#********************************************************************************
    
## Step 4: Importing the Recurrent Neural Network RNN Model

#********************************************************************************

regressor = Sequential()

regressor.add(LSTM(units = 50, return_sequences = True, input_shape = (vAR_X_train.shape[1], 1)))

regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 50, return_sequences = True))

regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 50, return_sequences = True))

regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 50))

regressor.add(Dropout(0.2))

regressor.add(Dense(units = 1))

regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')


#********************************************************************************
    
## Step 5: Training the Model

#********************************************************************************

#regressor.fit(vAR_X_train, vAR_y_train, epochs = 1, batch_size = 32)

#********************************************************************************

## Step 6: Import the Test Data

#********************************************************************************

vAR_TEST_DATA = vAR_pd.read_excel(vAR_Test_Data_Path)

vAR_TEST_DATA.head()

vAR_TEST_DATA = vAR_TEST_DATA.iloc[:,7]

#vAR_TRAINING_DATA = vAR_pd.read_excel(vAR_Training_Data,sheet_name=vAR_TRAIN)

vAR_dataset_total = vAR_pd.concat((vAR_TRAINING_DATA.iloc[:,8], vAR_TEST_DATA), axis = 0)

vAR_inputs = vAR_dataset_total[len(vAR_dataset_total) - len(vAR_TEST_DATA) - 60:].values

vAR_inputs = vAR_inputs.reshape(-1,1)

vAR_inputs = vAR_scaler.transform(vAR_inputs)

vAR_X_test = []

for i in range(60, 6540):
    
    vAR_X_test.append(vAR_inputs[i-60:i, 0])
    
vAR_X_test = vAR_np.array(vAR_X_test)

vAR_X_test = vAR_np.reshape(vAR_X_test, (vAR_X_test.shape[0], vAR_X_test.shape[1], 1))


#********************************************************************************

## Step 7: Run Forecasting

#********************************************************************************

vAR_Predicted_Energy_Demand = regressor.predict(vAR_X_test)

vAR_Predicted_Energy_Demand = vAR_scaler.inverse_transform(vAR_Predicted_Energy_Demand)

vAR_Predicted_Energy_Demand = vAR_pd.DataFrame(vAR_Predicted_Energy_Demand)

vAR_Predicted_Energy_Demand['Forecasted Energy Demand in MW'] = vAR_Predicted_Energy_Demand

vAR_Predicted_Energy_Demand = vAR_Predicted_Energy_Demand.iloc[:,1]

vAR_Predicted_Energy_Demand

vAR_TEST_DATA = vAR_pd.read_excel(vAR_Test_Data_Path)

vAR_TEST_DATA['Forecasted Energy Demand in MW'] = vAR_Predicted_Energy_Demand

vAR_TEST_DATA.head()

vAR_MODEL_OUTCOME_FORECASTED_ENERGY_DEMAND = vAR_TEST_DATA.to_excel(vAR_Model_Outcome1,index=False)

vAR_MODEL_OUTCOME_FORECASTED_ENERGY_DEMAND = vAR_pd.read_excel(vAR_Model_Outcome1)

vAR_MODEL_OUTCOME_FORECASTED_ENERGY_DEMAND.head()
```

# 6 Months Rolling Forecast

```python
vAR_FORECAST_ENERGY_DEMAND = vAR_pd.read_excel(vAR_Model_Outcome1)

vAR_ROLLING_FORECAST = vAR_FORECAST_ENERGY_DEMAND[vAR_FORECAST_ENERGY_DEMAND['Year']==2021]

vAR_ROLLING_FORECAST_MAY = vAR_ROLLING_FORECAST[vAR_FORECAST_ENERGY_DEMAND['Month']=='May']

vAR_ROLLING_FORECAST_MAY.reset_index()

vAR_ROLLING_FORECAST_JUN = vAR_ROLLING_FORECAST[vAR_FORECAST_ENERGY_DEMAND['Month']=='Jun']

vAR_ROLLING_FORECAST_JUN.reset_index()

vAR_ROLLING_FORECAST_JUL = vAR_ROLLING_FORECAST[vAR_FORECAST_ENERGY_DEMAND['Month']=='Jul']

vAR_ROLLING_FORECAST_AUG = vAR_ROLLING_FORECAST[vAR_FORECAST_ENERGY_DEMAND['Month']=='Aug']

vAR_ROLLING_FORECAST_SEP = vAR_ROLLING_FORECAST[vAR_FORECAST_ENERGY_DEMAND['Month']=='Sep']

vAR_ROLLING_FORECAST_OCT = vAR_ROLLING_FORECAST[vAR_FORECAST_ENERGY_DEMAND['Month']=='Oct']

vAR_ROLLING_FORECAST_FINAL = vAR_pd.concat([vAR_ROLLING_FORECAST_MAY,vAR_ROLLING_FORECAST_JUN,vAR_ROLLING_FORECAST_JUL,vAR_ROLLING_FORECAST_AUG,vAR_ROLLING_FORECAST_SEP,vAR_ROLLING_FORECAST_OCT])

vAR_ROLLING_FORECAST_FINAL

vAR_ROLLING_FORECAST = vAR_ROLLING_FORECAST_FINAL.to_excel(vAR_Model_Outcome_6Months_Rolling_Forecast,index=False)

vAR_ROLLING_FORECAST = vAR_pd.read_excel(vAR_Model_Outcome_6Months_Rolling_Forecast)

vAR_ROLLING_FORECAST.head()
```
