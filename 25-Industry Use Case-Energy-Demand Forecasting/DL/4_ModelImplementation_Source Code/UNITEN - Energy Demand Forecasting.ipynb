# ENERGY DEMAND FORECASTING USING S-ARIMA

```python
#********************************************************************************
    
## Step 1: Importing the Required Libraries

#********************************************************************************

import os

import yaml

vAR_yaml_file_path = 'C:/AI/Energy Demand Forecasting/ML/YAML/'

for vAR_path in os.listdir(vAR_yaml_file_path):
   
    vAR_final_path = os.path.join(vAR_yaml_file_path,'ENERGY_DEMAND_FORECASTING.yaml')
   
    with open (vAR_final_path) as vAR_file:
      
      config = yaml.safe_load(vAR_file)

## Reading the Training Data Path

vAR_Training_Data_Path = config.get("TRAINING_DATA_PATH")
   
#print(vAR_Training_Data_Path)

vAR_Test_Data_Path = config.get("TEST_DATA_PATH")

#print(vAR_Test_Data_Path)

vAR_Model_Outcome = config.get("MODEL_OUTCOME")

#print(vAR_Model_Outcome)

vAR_Model_Outcome1 = config.get("MODEL_OUTCOME_ALL_DIMENSIONS")

#print(vAR_Model_Outcome1)

vAR_Model_Outcome_6Months_Rolling_Forecast = config.get("MODEL_OUTCOME_6MONTHS_ROLLLING_FORECAST")

#print(vAR_Model_Outcome_6Months_Rolling_Forecast)

#********************************************************************************
    
## Step 2: Importing the Required Libraries

#********************************************************************************

import configparser

import os

import numpy as vAR_np

import pandas as vAR_pd

import matplotlib.pylab as vAR_plt

from datetime import datetime

%matplotlib inline

from matplotlib.pylab import rcParams

import statsmodels.api as sm

rcParams['figure.figsize'] = 12,9


#********************************************************************************
    
## Step 3: Importing the Training Data

#********************************************************************************

vAR_INPUT_DATA = vAR_pd.read_excel(vAR_Training_Data_Path)

#vAR_INPUT_DATA.head()

vAR_INPUT_DATA1 = vAR_pd.DataFrame(vAR_INPUT_DATA)

vAR_INPUT_DATA['Time'] = vAR_pd.to_datetime(vAR_INPUT_DATA['Time'])

vAR_INPUT_DATA['Time'] = vAR_pd.to_datetime(vAR_INPUT_DATA['Time'])

vAR_INPUT_DATA = vAR_INPUT_DATA[['Time', 'Region', 'County', 'State', 'City', 'Year', 'Month','Energy Consumption in MW']]

vAR_TRAINING_DATA = vAR_INPUT_DATA.set_index(['Time', 'Region', 'County', 'State', 'City', 'Year', 'Month'])

#vAR_TRAINING_DATA.head()

#********************************************************************************
    
## Step 4: Importing the SARIMAX Model

#********************************************************************************

import statsmodels.api as sm

#vAR_IndexedDataset
    
vAR_mod = sm.tsa.statespace.SARIMAX(vAR_TRAINING_DATA, order=(0, 1, 1),seasonal_order=(1, 1, 0, 12),enforce_stationarity=False,enforce_invertibility=False)


#********************************************************************************
    
## Step 5: Training the Model

#********************************************************************************

vAR_results = vAR_mod.fit()

vAR_pred = vAR_results.get_prediction(steps=60)

#vAR_pred

State = ['Washington']

City = ['Olympia', 'Seattle', 'Sacramento', 'Los Angeles']

vAR_APPEND_FORCAST_DF = vAR_pd.DataFrame(columns=['Forecasted Energy Demand in KW'])

for i in (State):
        
    for j in (City):
                                 

        #********************************************************************************
    
        ## Step 7: Run Forecasting

        #********************************************************************************
        
        vAR_results_data = vAR_results.get_forecast(steps=60).predicted_mean.astype(int)

        vAR_results_data_df = vAR_pd.DataFrame(vAR_results_data,columns=['Forecasted Energy Demand in MW'])
            
        vAR_RESULT1s_data_df = vAR_pd.DataFrame(vAR_results_data_df,columns=['Forecasted Energy Demand in MW'])
            
        vAR_APPEND_FORCAST_DF = vAR_APPEND_FORCAST_DF.append(vAR_results_data_df)
                
        vAR_Non_IndexedDataset = vAR_APPEND_FORCAST_DF.reset_index() 
                
        vAR_results_data_df = vAR_Non_IndexedDataset 

        vAR_results_data_df = vAR_results_data_df['Forecasted Energy Demand in MW']
                
        #vAR_results_data_df
            
            
        #********************************************************************************
    
        ## Step 8: Write the Model Outcome to a file

        #********************************************************************************
                
        vAR_results_data_df.to_excel(vAR_Model_Outcome,index=False)
            
        vAR_Prediction = vAR_results.get_prediction(steps=60).predicted_mean.astype(int)

            
#********************************************************************************
    
## Step 6: Import the Test Data

#********************************************************************************
            
vAR_test_data = vAR_pd.read_excel(vAR_Test_Data_Path)

vAR_test_data_for_loop = vAR_pd.read_excel(vAR_Model_Outcome)

vAR_test_data_for_loop = vAR_test_data_for_loop[:]
                    
vAR_test_data_for_loop

vAR_test_data = vAR_test_data.merge(vAR_test_data_for_loop,left_index=True,right_index=True)

vAR_test_data.head()

vAR_test_data = vAR_test_data[['Region', 'County', 'State', 'City', 'Year', 'Month','Time','Forecasted Energy Demand in MW']]

vAR_test_data = vAR_test_data.to_excel(vAR_Model_Outcome1,index=False) 

vAR_test_data = vAR_pd.read_excel(vAR_Model_Outcome1)

vAR_test_data.head()
```

# 6 Months Rolling Energy Demand Forecast

```python
vAR_FORECAST_ENERGY_DEMAND = vAR_pd.read_excel(vAR_Model_Outcome1)

vAR_ROLLING_FORECAST = vAR_FORECAST_ENERGY_DEMAND[vAR_FORECAST_ENERGY_DEMAND['Year']==2021]

vAR_ROLLING_FORECAST_MAY = vAR_ROLLING_FORECAST[vAR_FORECAST_ENERGY_DEMAND['Month']=='May']

vAR_ROLLING_FORECAST_MAY.reset_index()

vAR_ROLLING_FORECAST_JUN = vAR_ROLLING_FORECAST[vAR_FORECAST_ENERGY_DEMAND['Month']=='Jun']

vAR_ROLLING_FORECAST_JUN.reset_index()

vAR_ROLLING_FORECAST_JUL = vAR_ROLLING_FORECAST[vAR_FORECAST_ENERGY_DEMAND['Month']=='Jul']

vAR_ROLLING_FORECAST_AUG = vAR_ROLLING_FORECAST[vAR_FORECAST_ENERGY_DEMAND['Month']=='Aug']

vAR_ROLLING_FORECAST_SEP = vAR_ROLLING_FORECAST[vAR_FORECAST_ENERGY_DEMAND['Month']=='Sep']

vAR_ROLLING_FORECAST_OCT = vAR_ROLLING_FORECAST[vAR_FORECAST_ENERGY_DEMAND['Month']=='Oct']

vAR_ROLLING_FORECAST_FINAL = vAR_pd.concat([vAR_ROLLING_FORECAST_MAY,vAR_ROLLING_FORECAST_JUN,vAR_ROLLING_FORECAST_JUL,vAR_ROLLING_FORECAST_AUG,vAR_ROLLING_FORECAST_SEP,vAR_ROLLING_FORECAST_OCT])

vAR_ROLLING_FORECAST_FINAL

vAR_ROLLING_FORECAST = vAR_ROLLING_FORECAST_FINAL.to_excel(vAR_Model_Outcome_6Months_Rolling_Forecast,index=False)

vAR_ROLLING_FORECAST = vAR_pd.read_excel(vAR_Model_Outcome_6Months_Rolling_Forecast)

vAR_ROLLING_FORECAST.head()
```
