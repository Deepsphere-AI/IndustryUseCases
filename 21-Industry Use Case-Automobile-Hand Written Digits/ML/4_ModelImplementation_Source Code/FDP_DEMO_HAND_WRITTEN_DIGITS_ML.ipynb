```python
#********************************************************************************

## Step 1: INI File Configuration

#********************************************************************************

import configparser

import os

vAR_Config = configparser.ConfigParser(allow_no_value=True)

vAR_INI_FILE_PATH = os.getenv('HAND_WRITTEN_DIGITS_INI_ML')

#print(vAR_INI_FILE_PATH)

vAR_Config.read(vAR_INI_FILE_PATH)

vAR_Data = vAR_Config.sections()

vAR_Config.sections()

vAR_Training_Data = vAR_Config['FILE PATH']['TRAINING_DATA']
#print(vAR_Training_Data)

vAR_Training_Data_Excel_Sheet = vAR_Config['FILE PATH']['TRAINING_DATA_EXCEL_WORKSHEET']
#print(vAR_Training_Data_Excel_Sheet)

vAR_Test_Data = vAR_Config['FILE PATH']['TEST_DATA']
#print(vAR_Test_Data)

vAR_Test_Data_Excel_Sheet = vAR_Config['FILE PATH']['TEST_DATA_EXCEL_WORKSHEET']
#print(vAR_Test_Data_Excel_Sheet)

vAR_Model_Outcome = vAR_Config['FILE PATH']['MODEL_OUTCOME']
#print(vAR_Model_Outcome)

#********************************************************************************
    
## Step 2: Importing the Required Libraries

#********************************************************************************

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier

#********************************************************************************
    
## Step 3: Importing the Training Data

#********************************************************************************

vAR_TRAINING_DATA = pd.read_csv(vAR_Training_Data)


#********************************************************************************
    
## Step 4: Importing the Decision Tree Model

vAR_model = DecisionTreeClassifier(max_depth = 2)


#********************************************************************************
    
## Step 5: Training the Model

vAR_TRAINING_DATA_FEATURES = vAR_TRAINING_DATA.iloc[:,1:]

vAR_TRAINING_DATA_LABEL = vAR_TRAINING_DATA.iloc[:,0]

vAR_dtree_model_Training = vAR_model.fit(vAR_TRAINING_DATA_FEATURES, vAR_TRAINING_DATA_LABEL)


#********************************************************************************
    
## Step 6: Import the Test Data

#********************************************************************************

vAR_TEST_DATA = pd.read_csv(vAR_Test_Data)

vAR_TEST_DATA_FEATURES = vAR_TRAINING_DATA.iloc[:,1:]

#********************************************************************************
    
## Step 7: Hand Written Digits Classification 

#********************************************************************************

vAR_dtree_model_Test = vAR_model.predict(vAR_TEST_DATA_FEATURES)

vAR_dtree_model_Test = pd.DataFrame(vAR_dtree_model_Test,columns=['Predicted Digits'])

#vAR_dtree_model_Test

vAR_TEST_DATA = vAR_TEST_DATA.iloc[:,1:]

vAR_TEST_DATA['Predicted Digits'] = vAR_dtree_model_Test

vAR_TEST_DATA.head()

#********************************************************************************
```

```python
#********************************************************************************
    
## Step 8: Write the Model Outcome to a file

#********************************************************************************

vAR_TEST_DATA_OUTCOME = vAR_TEST_DATA.to_excel(vAR_Model_Outcome,index=False)

vAR_TEST_DATA_OUTCOME = pd.read_excel(vAR_Model_Outcome)

vAR_TEST_DATA_OUTCOME.head()

#********************************************************************************
```
