```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('airline cost new.csv')
```

```python
data
```

```python
X=data['num_pass']
```

```python
y=data['cost']
```

```python
plt.scatter(X,y)
plt.xlabel('number of passengers')
plt.ylabel('cost')
```

```python
import statsmodels.api as sm
```

```python
Xc=sm.add_constant(X)
results=sm.OLS(y,Xc).fit()
```

```python
results.summary()
```
