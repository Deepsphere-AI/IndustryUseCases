```python
#********************************************************************************

## Step 1: INI File Configuration

#********************************************************************************

import configparser
import os
vAR_Config = configparser.ConfigParser(allow_no_value=True)

vAR_INI_FILE_PATH = os.getenv('GENDER_DETECTION_INI_FILE_ML')
#print(vAR_INI_FILE_PATH)

vAR_Config.read(vAR_INI_FILE_PATH)

vAR_Data = vAR_Config.sections()

vAR_Config.sections()

vAR_Training_Data = vAR_Config['FILE PATH']['TRAINING_DATA']
#print(vAR_Training_Data)

vAR_Training_Data_Excel_Sheet = vAR_Config['FILE PATH']['TRAINING_DATA_EXCEL_WORKSHEET']
#print(vAR_Training_Data_Excel_Sheet)

vAR_Test_Data = vAR_Config['FILE PATH']['TEST_DATA']
#print(vAR_Test_Data)

vAR_Test_Data_Excel_Sheet = vAR_Config['FILE PATH']['TEST_DATA_EXCEL_WORKSHEET']
#print(vAR_Test_Data_Excel_Sheet)

vAR_Model_Outcome = vAR_Config['FILE PATH']['MODEL_OUTCOME']
#print(vAR_Model_Outcome)

vAR_Feature1 = vAR_Config['FEATURES']['FEATURE1']
#print(vAR_Feature1)

vAR_Feature2 = vAR_Config['FEATURES']['FEATURE2']
#print(vAR_Feature2)

vAR_Feature3 = vAR_Config['FEATURES']['FEATURE3']
#print(vAR_Feature3)

vAR_Feature4 = vAR_Config['FEATURES']['FEATURE4']
#print(vAR_Feature4)

vAR_Feature5 = vAR_Config['FEATURES']['FEATURE5']
#print(vAR_Feature5)

vAR_Feature6 = vAR_Config['FEATURES']['FEATURE6']
#print(vAR_Feature6)

vAR_Feature7 = vAR_Config['FEATURES']['FEATURE7']
#print(vAR_Feature7)

vAR_Label = vAR_Config['LABEL']['LABEL']
#print(vAR_Label)


#********************************************************************************
    
## Step 2: Importing the Required Libraries

#********************************************************************************

import pandas as vAR_pd

import numpy as vAR_np

from sklearn.linear_model import LogisticRegression

from sklearn.preprocessing import LabelEncoder


#********************************************************************************
    
## Step 3: Importing the Training Data

#********************************************************************************

vAR_INPUT_DATA = vAR_pd.read_excel(vAR_Training_Data,sheet_name=vAR_Training_Data_Excel_Sheet)

vAR_INPUT_DATA = vAR_pd.DataFrame(vAR_INPUT_DATA)

vAR_INPUT_DATA.head()

vAR_TRAINING_DATA = vAR_INPUT_DATA

vAR_le = LabelEncoder()


#********************************************************************************

vAR_Appearance_Conversion = vAR_le.fit_transform(vAR_TRAINING_DATA.iloc[:,6])

vAR_Eye_Conversion = vAR_le.fit_transform(vAR_TRAINING_DATA.iloc[:,7])

vAR_Hair_Conversion = vAR_le.fit_transform(vAR_TRAINING_DATA.iloc[:,8])

vAR_Skin_Colour_Conversion = vAR_le.fit_transform(vAR_TRAINING_DATA.iloc[:,9])

vAR_Gender_Conversion = vAR_le.fit_transform(vAR_TRAINING_DATA.iloc[:,16])

#********************************************************************************


vAR_df1 = vAR_pd.DataFrame(columns=['First Name','Last Name','Age','Height','Weight','Feet Size','Appearance','Eye','Hair','Skin Colour','Nationality','Education','Occupation','Maritial Status','Phone Number','Email','Gender'])

vAR_df1[vAR_Feature1] = vAR_Appearance_Conversion

vAR_df1[vAR_Feature2] = vAR_Eye_Conversion

vAR_df1[vAR_Feature3] = vAR_Hair_Conversion

vAR_df1[vAR_Feature4] = vAR_Skin_Colour_Conversion

vAR_df1[vAR_Label] = vAR_Gender_Conversion


vAR_TRAININNG_DATA_FEATURES = vAR_df1[[vAR_Feature1, vAR_Feature2, vAR_Feature3, vAR_Feature4]]

vAR_TRAININNG_DATA_LABEL = vAR_df1[vAR_Label]

#********************************************************************************
    
## Step 4: Importing the Logistic Regression Model

vAR_MODEL = LogisticRegression()


#********************************************************************************
    
## Step 5: Training the Model

vAR_MODEL_TRAINING = vAR_MODEL.fit(vAR_TRAININNG_DATA_FEATURES,vAR_TRAININNG_DATA_LABEL)


#********************************************************************************
    
## Step 6: Import the Test Data

#********************************************************************************

vAR_TEST_DATA = vAR_pd.read_excel(vAR_Test_Data,sheet_name=vAR_Test_Data_Excel_Sheet)

vAR_TEST_DATA = vAR_pd.DataFrame(vAR_TEST_DATA)

vAR_TEST_DATA.head()

#********************************************************************************

vAR_Appearance_Conversion = vAR_le.fit_transform(vAR_TEST_DATA.iloc[:,6])

vAR_Eye_Conversion = vAR_le.fit_transform(vAR_TEST_DATA.iloc[:,7])

vAR_Hair_Conversion = vAR_le.fit_transform(vAR_TEST_DATA.iloc[:,8])

vAR_Skin_Colour_Conversion = vAR_le.fit_transform(vAR_TEST_DATA.iloc[:,9])

#********************************************************************************

vAR_df2 = vAR_pd.DataFrame(columns=['First Name','Last Name','Age','Height','Weight','Feet Size','Appearance','Eye','Hair','Skin Colour','Nationality','Education','Occupation','Maritial Status','Phone Number','Email'])

#vAR_df2 = vAR_pd.DataFrame(columns=[vAR_Feature1, vAR_Feature2, vAR_Feature3, vAR_Feature4, vAR_Feature5, vAR_Feature6, vAR_Feature8, vAR_Feature9, vAR_Feature10, vAR_Feature11])

vAR_df2[vAR_Feature1] = vAR_Appearance_Conversion

vAR_df2[vAR_Feature2] = vAR_Eye_Conversion

vAR_df2[vAR_Feature3] = vAR_Hair_Conversion

vAR_df2[vAR_Feature4] = vAR_Skin_Colour_Conversion


vAR_TEST_DATA_FEATURES = vAR_df2[[vAR_Feature1, vAR_Feature2, vAR_Feature3, vAR_Feature4]]

vAR_TEST_DATA_FEATURES

#********************************************************************************
    
## Step 7: Gender Detection (Model Prediction)

#********************************************************************************

vAR_MODEL_TEST = vAR_MODEL.predict(vAR_TEST_DATA_FEATURES)

vAR_MODEL_TEST_GENDER = vAR_pd.DataFrame(vAR_MODEL_TEST,columns=['Predicted Gender'])

vAR_TEST_DATA['Predicted Gender'] = vAR_MODEL_TEST_GENDER

vAR_TEST_DATA['Predicted Gender'] = vAR_TEST_DATA['Predicted Gender'].replace({0: "Male"})

vAR_TEST_DATA['Predicted Gender'] = vAR_TEST_DATA['Predicted Gender'].replace({1: "Female"})

vAR_MODEL_OUTCOME_GENDER = vAR_TEST_DATA

vAR_MODEL_OUTCOME_GENDER


#********************************************************************************
    
## Step 8: Write the Model Outcome to a file

#********************************************************************************

vAR_TEST_DATA = vAR_TEST_DATA.to_excel(vAR_Model_Outcome,index=False) 

vAR_TEST_DATA = vAR_pd.read_excel(vAR_Model_Outcome)

vAR_TEST_DATA

#********************************************************************************
```
