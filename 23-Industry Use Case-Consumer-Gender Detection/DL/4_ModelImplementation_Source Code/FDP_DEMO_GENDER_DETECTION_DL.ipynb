```python
def process_test_data():
    
    vAR_testing_data=[]
    
    for vAR_img in tqdm(os.listdir(vAR_Training_Data)):
    
        vAR_path=os.path.join(vAR_Training_Data,vAR_img)
        
        vAR_img_num=vAR_img.split('_')[-1]
        
        vAR_img1=cv2.resize(cv2.imread(vAR_path, cv2.IMREAD_GRAYSCALE), (vAR_Image_Size, vAR_Image_Size))
        
        vAR_testing_data.append([vAR_np.array(vAR_img1), vAR_img_num])
    
    vAR_np.save('vAR_test_data.npy', vAR_testing_data)
    
    return vAR_testing_data

vAR_test_data = process_test_data()


for num, data in enumerate(vAR_test_data[:]):
    
    vAR_img_num=data[1]
    
    vAR_img_data=data[0]
    
    data=vAR_img_data.reshape(vAR_Image_Size,vAR_Image_Size,1)
    
    vAR_model_out=model.predict([data])[0]
    
    print(vAR_model_out)
    
    if vAR_np.argmax(vAR_model_out)==0 : 
        
        vAR_str_label="Male"
        
        print(vAR_img_num,vAR_str_label)
            
    else: 
                
        vAR_str_label='Female'
        
        print(vAR_img_num,vAR_str_label)
```
